#### За допомогою графічного редактора визначте точки відповідності (кути дошки) на зображеннях chess_1 та chess_2, що додані до репозиторію лекції. Проведіть гомографічне перетворення для приведення зображення chess_2 до вигляду базового зображення chess_1.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
im_dst = cv2.imread('chess_1.png')
pts_dst = np.array([[464, 39],[460, 261],[200, 245],[190, 50]])
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.axis('off')
plt.show()
```

```python
im_src = cv2.imread("chess_2.png")
pts_src = np.array([[166, 33], [530, 100], [409, 364],[189, 317]])
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.axis('off')
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)

im_out = cv2.warpPerspective(im_src,
                             h,
                             (im_dst.shape[1], im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.axis('off')
plt.show();
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 15), sharex=True, sharey=True)

images = [im_src, im_dst, im_out]
titles = ['chess_2', 'chess_1', 'Гомографічне перетворення chess_2 до chess_1']

for ax, img, title in zip(axs, images, titles):
    ax.axis('off')
    ax.imshow(img, cmap='gray')
    ax.set_title(title)

plt.show()
```
